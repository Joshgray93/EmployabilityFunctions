# Radar Charts for Employability Data

- Creates a radar chart and plots a comparison indicator as well, the comparison indicator is the same for all the interations of the chart.
- All data used is illustrative and not representative of the schools

## Example of Radar Chart to Create

- Below is an eg of a radar chart like the one we intend to create

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')

categories = ['Engagement Overall', 'Career Confidence', 'Career Planning', 'Emp. Experience', 'Graduate Outcomes']
categories = [*categories, categories[0]]

DEV = [4, 1, 2, 4, 5]
UEA = [3, 2, 3, 4, 3]
DEV = [*DEV, DEV[0]]
UEA = [*UEA, UEA[0]]

label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(DEV))
plt.figure(figsize=(8, 8))
plt.subplot(polar=True)
plt.plot(label_loc, DEV, color='coral', lw=8, marker='o', markerfacecolor='red', markeredgecolor='black',
        label='EDU')
plt.plot(label_loc, UEA, label='UEA', lw=4, color='grey', ls='--')
plt.title('EDU - Employability Index', size=20, y=1.05)
lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories, size=12)
plt.legend()
plt.show()
```
