# Line Plot Template

```python
'''Set fig and dpi'''
fig = plt.figure(figsize=(18, 6), dpi=400)
ax = fig.gca()
fig.patch.set_facecolor('#f6f6f6')

'''Plot data, add 2nd plot if eneded for track line with grey as the colour, separate into two dfs based on year for this'''

ax.plot(df['Week'], df['Count'], color='#71C1ED', lw=4, marker='o', label='Data1')

'''Set the integer'''
plt.xticks(np.arange(1, 23, 2))
plt.yticks(np.arange(0, 3000, 500))

'''Add background color'''
ax.set_facecolor('#ffffff')

'''Label the plot'''
plt.xlabel("Add your xlabel", fontfamily='Arial', size=18, color='dimgray')
plt.ylabel("Add your y label", fontfamily='Arial', size=18, color='dimgray')
plt.title("Title Can Run Over Two Lines\nSecond Line Text", fontfamily='Arial', size=30, 
          loc='left', pad=20, color='#36454F')


'''Add tick styles'''
plt.xticks(fontsize=12, fontfamily='Arial', color='#36454F')
plt.yticks(fontsize=12, fontfamily='Arial', color='#36454F')
plt.xlim([0.5,22.5])

'''Add axis design'''
ax.spines[['right', 'top']].set_visible(False)
ax.spines[['bottom', 'left']].set_visible(True)
ax.spines['bottom'].set_color('#36454F')
ax.spines['left'].set_color('#36454F')

'''Add data notes'''
plt.text(0.5,-550, '___'*30, color='#36454F', size=20, fontfamily='Arial', fontweight='bold')
plt.text(0.5,-800, 'Include data note if needed', color='#36454F', size=20, fontfamily='Arial')
plt.text(0.5,-1000, 'Include a second data note if needed', color='#36454F', size=20, fontfamily='Arial')


'''Set label positioning'''
ax.xaxis.set_label_coords(.5, -.09)
ax.yaxis.set_label_coords(-0.05, 0.5)

'''Add grid'''
plt.grid(alpha=0.2, color='#36454F')


'''Add legend'''
plt.legend(loc='center left', bbox_to_anchor=(0.895, 1.1), fontsize=16)

'''Add in your annotation if neeeded'''
ax.annotate('Annotation', color='#0D5F7D', size=18, xy=(8.9, 1500),  xycoords='data',
            xytext=(-50, 10), textcoords='offset points',
            arrowprops=dict(arrowstyle='->', facecolor='#0D5F7D', ec='#0D5F7D'),
            horizontalalignment='right', verticalalignment='top')
            
    
'''Show plot'''
plt.show()
```
